```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/DatumLearning/Machine-Learning-Playlist/main/AAPL.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
for col in df.columns:
  if df[col].dtype == 'object':
    print(col , df[col].dtype)
```

```python
df.drop(["Date" , "Date_col"] , axis = 1 , inplace = True)
```

```python
df.head()
```

```python
features = df.drop("Close_forcast" , axis = 1)
target = df.Close_forcast
```

```python
features.head()
```

```python
target
```

```python
rows = df.shape[0]
```

```python
train_size = int(0.88 * rows)
val_size = int(0.1 * rows)
test_size = int(0.02 * rows)
```

```python
train_size , val_size , test_size
```

```python
train_inds = train_size
val_inds = train_inds + val_size
test_inds = val_inds + test_size
```

```python
train_inds , val_inds , test_inds
```

```python
train_x , val_x , test_x = features[:train_inds] , features[train_inds : val_inds] , features[val_inds:test_inds]
train_y , val_y , test_y = target[:train_inds] , target[train_inds : val_inds] , target[val_inds:test_inds]
```

```python
train_x.shape
```

```python
lr = LinearRegression()
```

```python
lr.fit(train_x , train_y)
```

```python
preds_train = lr.predict(train_x)
preds_val = lr.predict(val_x)
preds_test = lr.predict(test_x)
```

```python
print("Training error : " , mean_squared_error(train_y , preds_train))
```

```python
print("validation error : " , mean_squared_error(val_y , preds_val))
```

```python
print("Testing error : " , mean_squared_error(test_y , preds_test))
```
